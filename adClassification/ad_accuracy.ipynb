# 정확도 계산 코드

```python
class Accuracy:
  def __init__(self, correct, predicted):
    mat = confusion_matrix(correct, predicted)
    self.tn = mat[0][0]
    self.fp = mat[0][1]
    self.fn = mat[1][0]
    self.tp = mat[1][1]

    print('tp:', self.tp)
    print('fp:', self.fp)
    print('tn:', self.tn)
    print('fn:', self.fn)

  def calc_accuracy(self):
    return (correct_data['ad'] == kobert_data['ad']).sum() / len(kobert_data['ad'])

  # 학습결과가 정답과 같을 확률
  def calc_precision(self):
    return self.tp / (self.tp + self.fp)
    
  # 실제 광고를 광고로 분류할 확률
  def calc_recall(self):
    return self.tp / (self.tp + self.fn)

  def calc_f_measure(self):
    precision = self.calc_precision()
    recall =  self.calc_recall()
    return 2 * (precision * recall) / (precision + recall)
```

# 정확도 계산

```python
dir_path = '/content/drive/MyDrive/forColab/final/ilsan_pizza/summary/'

correct_answer = dir_path + 'correct_answer.csv'
kobert_answer = dir_path + 'kobert_answer.csv'
```

```python
correct_data = pd.read_csv(correct_answer)
kobert_data = pd.read_csv(kobert_answer)
```

```python
correct_data_ad = correct_data[correct_data['ad']==True]
correct_data_ad['ad'] = '0'
correct_data_non_ad = correct_data[correct_data['ad']==False]
correct_data_non_ad['ad'] = '1'

correct_data = pd.concat([correct_data_ad, correct_data_non_ad]).sort_index()

kobert_data_ad = kobert_data[kobert_data['ad']==True]
kobert_data_ad['ad'] = '0'
kobert_data_non_ad = kobert_data[kobert_data['ad']==False]
kobert_data_non_ad['ad'] = '1'

kobert_data = pd.concat([kobert_data_ad, kobert_data_non_ad]).sort_index()
```

```python
acc = Accuracy(correct_data['ad'], kobert_data['ad'])

print('\nprediction performances (0~1)')
print('accuracy:', acc.calc_accuracy())
print('precision:', acc.calc_precision())
print('recall:', acc.calc_recall())
print('f_measure:', acc.calc_f_measure())
```
